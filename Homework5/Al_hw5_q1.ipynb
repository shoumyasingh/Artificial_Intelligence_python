```python
import numpy as np
```

```python
data = np.array([[1,0,0,7,'Apple'],
                 [0,1,0,20,'Watermelon'],
                 [1,0,0,1,'Cherry'],
                 [0,1,0,7.5,'Apple'],
                 [1,0,0,1,'Strawberry'],
                 [1,0,0,0.8,'Cherry']])
```

```python
print(data)
```

```python
x = data[:,:4]
print(x)
```

```python
y = data[:,4]
print(y)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(bootstrap = 'true',max_features='sqrt',max_depth=4)
model.fit(x,y)
```

```python
test = np.array([[1,0,0,10]])
print(test)
```

```python
model.predict(test)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
estimator = model.estimators_[5]
plt.figure(figsize=(25,10))
a = plot_tree(estimator,
              feature_names= ['Red','Green','Blue','Size'],
              class_names = ['Apple','Watermelon','Cherry','Strawberry'],
              filled = True,
              rounded = True,
              fontsize = 14)
```
