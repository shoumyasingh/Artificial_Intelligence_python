```python
# In markov chain

import numpy as np

start_probability = [0.1, 0.6, 0.3]

transition_probability = np.matrix([
  [0.10, 0.70, 0.20],
  [0.75, 0.15, 0.10],
  [0.60, 0.35, 0.05]
])

p = start_probability * (transition_probability**3)
print(p)

# [P(London),P(Barcelona),P(New York)]

```
